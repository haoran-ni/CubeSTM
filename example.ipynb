```python
from cubestm import CubeSTM
from os.path import join
```

```python
dir = './data'
cube_path = join(dir, 'cube_001_stm_01.cube')
geo_path = join(dir, 'geometry.in')
instance = CubeSTM(cube_path, geo_path)
```

```python
instance.plot(show_atoms=['C','N'],
              z_decay=1.6,
              xy_decay=1.5,
              vmin=1e-4,
              vmax=0.4,
              inset_size=0.5,
              atoms_size=1.2,
              atoms_edge=1.5)
```

```python
instance.plot_cell(show_atoms=['C','N'],
                   supercell=[3,2],
                   z_decay=1.6,
                   xy_decay=1.5,
                   vmin=1e-4,
                   vmax=0.4,
                   atoms_size=1.2,
                   atoms_edge=1.5)
```
