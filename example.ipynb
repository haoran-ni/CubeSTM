```python
from cubestm import CubeSTM
from os.path import join
```

```python
dir = './data'
cube_path = join(dir, 'cube_001_stm_01.cube')
geo_path = join(dir, 'geometry.in')
instance = CubeSTM(cube_path, geo_path)
```

```python
instance.plot_canvas(show_atoms=['C','N'],
                     z_decay=1.6,
                     neighbourhood=4,
                     xy_decay=1.5,
                     vmin=1e-4,
                     vmax=0.4,
                     inset_size=0.45,
                     atoms_size=1.2,
                     atoms_edge=1.5)
```

```python
instance.plot_cell(supercell=[3,2],
                   z_decay=1.6,
                   neighbourhood=4,
                   xy_decay=1.5,
                   vmin=1e-4,
                   vmax=0.4)
```
